```python
def fibonacci(n):
   if n <= 1:
       return n
   else:
       return(fibonacci(n-1) + fibonacci(n-2))

n = int(input("Enter nth term till which you want the fibonacci series: "))

print("Fibonacci sequence:")
for i in range(n):
   print(fibonacci(i))
```

```python
def no_duplicates(li):
    new = []
    
    for i in li:
        if i not in new:
            new.append(i)
    return new

a = no_duplicates( [1,1,2,2,3,4,5,6,6] )
print(a)
```

```python
def no_duplicates(li):
    return list(set(li))

print( no_duplicates([1,1,2,2,3,4,5,6,6]) )
```

```python
def prime_num(n):
    if n <= 1:
        return "Not prime"
    else:
        
        for i in range(2, n//2):
            if n % i == 0:
                return "Not prime"
        else: 
            return "Prime"
        
n = int( input("Enter a number: ") )
print( prime_num(n) )
```
